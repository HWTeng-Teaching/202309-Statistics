```python
import pandas as pd
from IPython.display import display

# 假設您的資料框中有 '日期'、'數值' 兩個欄位
# 並且 '日期' 欄位已經被轉換為日期類型
# 如果沒有轉換，可以使用 df['日期'] = pd.to_datetime(df['日期']) 進行轉換

# 設定 '日期' 欄位為索引
df.set_index('日期', inplace=True)

# 指定您想要的時間範圍，這裡以2023年1月1日至2023年12月31日為例
start_date = '2023-01-01'
end_date = '2023-12-31'

# 選擇指定時間範圍內的資料
selected_data = df.loc[start_date:end_date]

# 計算平均值、標準差、中位數和四分位數
average_value = selected_data['淨值'].mean()
std_deviation = selected_data['淨值'].std()
median_value = selected_data['淨值'].median()
quartiles = selected_data['淨值'].quantile([0.25, 0.5, 0.75])

# 顯示結果
result_dict = {
    '平均值': average_value,
    '標準差': std_deviation,
    '中位數': median_value,
    '第一四分位數': quartiles[0.25],
    '中位數': quartiles[0.5],
    '第三四分位數': quartiles[0.75]
}

result_df = pd.DataFrame(result_dict, index=[0])
display(result_df)
```
